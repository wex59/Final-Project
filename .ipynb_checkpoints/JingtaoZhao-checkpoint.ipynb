```python
import pandas as pd
police = pd.read_csv("PittsburghPoliceArrestData.csv", usecols = ['INCIDENTNEIGHBORHOOD'])
police_dict = {}
i = 0
while i < 58322:
    item = police.loc[i,'INCIDENTNEIGHBORHOOD']
    if item in police_dict:
        police_dict[item]+=1
    else:
        police_dict[item]=1
    i += 1
final = pd.Series(police_dict, name = "place")
```

```python
final = final.sort_values()[:]
```
